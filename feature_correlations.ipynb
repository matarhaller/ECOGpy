## calculate difference between real latency and RT
* output to csv - mean, median and std of the difference

## add ROI

```python
SJdir = '/home/knight/matar/MATLAB/DATA/Avgusta/'
filename = os.path.join(SJdir, 'PCA','Stats', 'Correlations','maxes_vs_lats', 'maxes_vs_lats_corr.csv')
df = pd.read_csv(filename)
```

```python
filename = 'brain_ROI.mat'
filename = os.path.join(SJdir, 'ROIs', 'brain_ROI.mat')
data = loadmat.loadmat(filename)
brain_ROI = data['brain_ROI']
```

```python
def get_ROI(subj, e, brain_ROI, task = None):
    '''
    Returns an ROI for a given subj and elec.
    If subject is GP35 then must give a task argument so knows which ROIs to pull
    '''
    
    if (subj == 'GP35') & ((task == 'EmoRep') | (task == 'EmoGen')):
        subj = 'GP35_words'
    elif (subj == 'GP35') & ((task == 'FaceEmo') | (task == 'FaceGen')):
        subj = 'GP35_face'
    
    try:
        rois = brain_ROI[subj]
    except:
        return None
    else:
        for roi, elecs in rois.iteritems():
            if np.in1d(e, elecs):
                return roi
```

```python
df['ROI'] = df.apply(lambda x:get_ROI(x.subj, x.elec, brain_ROI, task = x.task), axis = 1)
```

```python
df = df[df.columns[1:]]
```

```python
filename = os.path.join(SJdir, 'PCA','Stats', 'Correlations','maxes_vs_lats', 'maxes_vs_lats_corr_withROI.csv')
df.to_csv(filename, index = False)
```

```python
import pandas as pd
from pandas.tools.plotting import scatter_matrix
import os
from __future__ import division
import numpy as np
import cPickle as pickle
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.preprocessing import scale
```

```python
SJdir = '/home/knight/matar/MATLAB/DATA/Avgusta/'
```

```python
subj ='GP15'
task = 'EmoGen'
```

```python
filename = os.path.join(SJdir, 'PCA', 'ShadePlots_hclust', 'elecs', 'significance_windows', 'data', ''.join([subj, '_', task, '.p']))
data_dict = pickle.load(open(filename, 'rb'))
```

```python
elecs = data_dict['means'].keys()
```

```python
features = ['means', 'maxes', 'lats_pro','stds']
features.append('RTs')
```

```python
big_dict = dict()
for f in features:
    filename = os.path.join(SJdir, 'PCA', 'ShadePlots_hclust', 'elecs', 'significance_windows', 'csv_files', ''.join([subj, '_', task, '_', f, '.csv']))
    df = pd.read_csv(filename)
    elecs = df.columns.values
    big_dict[f] = df
```

```python
big_dict['maxes_rel'] = big_dict['maxes']-big_dict['means']
```

```python
features.append('maxes_rel')
```

```python
to_scale = True
for e in elecs:
    if e!='5':
        continue
        
    if to_scale == True: #if to normalize features (easier to spot outliers)
        means, maxes, lats_pro, stds, RTs, maxes_rel = [scale(big_dict[f][e].astype(float)) for f in features]
    else:
        means, maxes, lats_pro, stds, RTs, maxes_rel = [big_dict[f][e] for f in features]
        
    tmp = pd.DataFrame({'means':means, 'maxes':maxes, 'maxes_rel':maxes_rel, 'lats_pro':lats_pro, 'stds':stds, 'RTs':RTs})

    f, ax = plt.subplots(figsize = (10, 10))
    scatter_matrix(tmp, ax = ax, grid = True, hist_kwds={'alpha': 0.5})
```

```python
#correlation each feature with RT and with features for each elec - values in csv for plot
flist = []
for f1 in features:
    for f2 in features:
        flist.append('_'.join([f1, f2]))

corr_dict = dict()
for e in elecs:
    corr_list = [subj, task]
    for f1 in features:
        for f2 in features:
            idx = np.where(~np.isnan(data_dict[f1][e]))
            corr_list.append(np.corrcoef(data_dict[f1][e][idx], data_dict[f2][e][idx])[0,1])
            corr_dict[e] = corr_list
```

```python
df = pd.DataFrame(corr_dict).T
df = df.reset_index()
```

```python
cols = ['elec','subj','task']
cols.extend(flist)
df.columns = cols
```
