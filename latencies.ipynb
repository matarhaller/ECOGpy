```python
import pandas as pd
import os
from __future__ import division
import numpy as np
import cPickle as pickle
import matplotlib.pyplot as plt
from scipy import stats
import loadmat
from pandas.tools.plotting import scatter_matrix
```

## plot single trial traces for an electrode
* mark significance window

```python
subj, task, elec = ('GP15', 'EmoGen', 19)

SJdir = '/home/knight/matar/MATLAB/DATA/Avgusta/'

filename = os.path.join(SJdir,'PCA', 'Stats', 'single_electrode_windows_withdesignation_EDITED.csv')
df = pd.read_csv(filename)

#load data
filename = os.path.join(SJdir, 'Subjs', subj, task, 'HG_elecMTX_percent.mat')
data_dict = loadmat.loadmat(filename)

active_elecs, Params, srate, RT, data_all = [data_dict.get(k) for k in ['active_elecs','Params','srate','RTs','data_percent']]
bl_st = Params['bl_st']
bl_st = bl_st/1000*srate

RT = RT + abs(bl_st) #RTs are calculated from stim onset, need to account for bl in HG_elecMTX_percent

row = df[(df.subj == subj) & (df.task == task) & (df.elec == elec)]
_, subj, task, cluster, pattern, elec, start_idx, end_idx, start_idx_resp, end_idx_resp, _, _ = row.values[0]

eidx = np.in1d(active_elecs, elec)
data = data_all[eidx,:,:].squeeze()

start_idx = start_idx + abs(bl_st)
end_idx = end_idx_resp + max(RT)

for i in range(data.shape[0]):
    f,ax = plt.subplots()
    ax.plot(np.arange(-500, data.shape[1]-500), data[i,:], label = None)
    ax.plot(np.arange(start_idx-500,end_idx-500), data[i,start_idx:end_idx], 'r', label = '%i - %i' %(start_idx-500, end_idx-500))
    ax.axvline(start_idx-500,color = 'k', label = None)
    ax.axvline(RT[i]-500,color = 'k', label = None)
    ax.set_title('%s %s %i - %s - trial %i' %(subj, task, elec, pattern, i))
    ax.legend()
    filename = os.path.join(SJdir, 'PCA','Stats','latencies', 'static','singletrials', '%s_%s_e%i_%i.png' %(subj, task, elec, i))
    plt.savefig(filename)
    plt.close()

plt.plot(data.T)
plt.plot(data.mean(axis = 0), 'k',lw = 4)
```
