```python
import brain_clusters
import os
import pandas as pd
import numpy as np
import pandas as pd
import scipy.io as spio
import matplotlib.gridspec as gridspec
import cPickle as pickle
import scipy.stats as stats
import matplotlib.pyplot as plt
import matplotlib
```

```python
SJdir = '/home/knight/matar/MATLAB/DATA/Avgusta'
reconlist = os.path.join(SJdir, 'PCA','reconlist.csv')
reconlist = pd.read_csv(reconlist)
```

```python
filename = os.path.join(SJdir, 'PCA', 'ShadePlots_hclust', 'elecs', 'significance_windows', 'smoothed', 'mean_traces', 'csv_files', 'mean_traces_all_subjs.csv')
df_onsets = pd.read_csv(filename)
```

```python
subj, task, param = ('GP15','EmoGen', 'means')
df_onsets[(df_onsets.subj == subj) & (df_onsets.task == task)].head()
```

```python
param, label, custom_cmap = ('means', '% change baseline', 'PiYG') #what to plot

saveDir = os.path.join(SJdir, 'PCA', 'ShadePlots_hclust', 'elecs', 'significance_windows', 'smoothed', 'mean_traces', 'images', param)
#os.mkdir(saveDir)

subj_task = df_onsets[['subj', 'task']].drop_duplicates()

for s_t in subj_task.itertuples():
    subj, task = s_t[1:]
    
    xycoords = reconlist[reconlist.subj_task == '_'.join([subj, task])].xyfile.iloc[0]
    if isinstance(xycoords, float): #no explicit xycoord path
        xycoords = 'xycoords.p'

    reconfile = reconlist[reconlist.subj_task == '_'.join([subj, task])].reconfile.iloc[0]
    if len(reconfile)>0:
        recon_filepath = os.path.join(SJdir, 'Subjs', subj, reconfile)
    else:
        print 'skipping %s %s' %(subj, task)
         #skip this subj if no reconfile

    #open xycoords dictionary                                                                                           
    filename = os.path.join(SJdir, 'Subjs',subj, xycoords)
    with open(filename, 'r') as f:
        xycoords = pickle.load(f)
        f.close()

    xycoords = pd.DataFrame(np.array(xycoords.values()), columns=['x_2d', 'y_2d'], index=np.array(xycoords.keys())+1)

    #custom_cmap = plt.get_cmap('bwr_r')
    #custom_cmap = plt.get_cmap('PiYG')
    custom_cmap = plt.get_cmap(custom_cmap)
    
    #get elecs of interest (for PC of interest)
    df = df_onsets[(df_onsets.subj == subj) & (df_onsets.task == task)]    

    weights = df[['elec',param]]
    weights.columns = ['elec','group']
    weights = weights.set_index('elec')

    #figure properties
    f, ax1 = plt.subplots(figsize = (10,10))
    ax1.set_title('%s %s %s' %(subj, task, param))

    #plot recon
    ax = brain_clusters.plot_xy_map(weights, locs = xycoords.loc[weights.index], ax = ax1, cmap = custom_cmap, szmult=200, cbar = False, im_path = os.path.join(SJdir,'Subjs', subj, reconfile))    
    cbar = ax.get_figure().colorbar(ax.collections[0], shrink = 0.75, label = label)

    filename = os.path.join(saveDir, '_'.join([subj, task, param + '.png']))
    plt.savefig(filename)
    plt.close('all')
```
