### plot latencies (lats) or onsets (start_idx) on brains
* calculated on mean trace (from ShadePlots_meanvalues)
* normalized or not
* duration electrodes in red gradient, stim/resp in blue green

```python
param, label, custom_cmap = ('start_idx', 'ms', 'Reds') #also for latencies
normalize = True #if to normalize values by median

saveDir = os.path.join(SJdir, 'PCA', 'ShadePlots_hclust', 'elecs', 'significance_windows', 'smoothed', 'mean_traces', 'images', param)
if normalize:
    saveDir = '_'.join([saveDir, 'norm_minRT'])
    label = ' '.join([label, 'norm_minRT'])
os.mkdir(saveDir)

subj_task = df_mean_values[['subj', 'task']].drop_duplicates()
    
for s_t in subj_task.itertuples():
    subj, task = s_t[1:]

    xycoords = reconlist[reconlist.subj_task == '_'.join([subj, task])].xyfile.iloc[0]
    if isinstance(xycoords, float): #no explicit xycoord path
        xycoords = 'xycoords.p'

    reconfile = reconlist[reconlist.subj_task == '_'.join([subj, task])].reconfile.iloc[0]
    if len(reconfile)>0:
        recon_filepath = os.path.join(SJdir, 'Subjs', subj, reconfile)
    else:
        print 'skipping %s %s' %(subj, task)
         #skip this subj if no reconfile

    #open xycoords dictionary                                                                                           
    filename = os.path.join(SJdir, 'Subjs',subj, xycoords)
    with open(filename, 'r') as f:
        xycoords = pickle.load(f)
        f.close()

    xycoords = pd.DataFrame(np.array(xycoords.values()), columns=['x_2d', 'y_2d'], index=np.array(xycoords.keys())+1)

    #get elecs of interest
    df = df_mean_values[(df_mean_values.subj == subj) & (df_mean_values.task == task)]
    #if normalize:
    #    df.lats = df.lats + df.start_idx #add start_idx back to latency so that relative to stim onset (not relevant for start_idx)
    
    weights = df[['elec',param]]
    weights.columns = ['elec','group']
    weights = weights.set_index('elec')

    #figure properties
    f, ax1 = plt.subplots(figsize = (10,10))
    ax1.set_title('%s %s %s' %(subj, task, param))
    if normalize:
        ax1.set_title('%s %s %s norm' %(subj, task, param))

    #plot recon
    if normalize:
        #vmin, vmax = (0, df.RTs_median.mean()) #normalize by median RT
        vmin, vmax = (0, df.RTs_min.mean()) #normalize by min RT
    else:    
        vmin = df[df.pattern == 'D'][param].min()
        vmax = df[df.pattern == 'D'][param].max()
        
    ax = brain_clusters.plot_xy_map(weights, locs = xycoords.loc[weights.index], ax = ax1, cmap = custom_cmap, szmult=200, cbar = True,  im_path = os.path.join(SJdir,'Subjs', subj, reconfile), vaxes = (vmin, vmax))    

    #overwrite resp and stim and sustained
    idx = (df.pattern == 'S') | (df.pattern == 'SR')
    x = xycoords.loc[df[idx].elec]['x_2d']
    y = xycoords.loc[df[idx].elec]['y_2d']
    tmp = ax.scatter(x, y, facecolor = 'lime', edgecolor = None,  s = 200)
    S = df[idx][param].mean()
    
    idx = (df.pattern == 'R')
    x = xycoords.loc[df[idx].elec]['x_2d']
    y = xycoords.loc[df[idx].elec]['y_2d']
    tmp = ax.scatter(x, y, facecolors = 'skyblue', edgecolor = None,  s = 200)
    R = df[idx][param].mean()

    idx = (df.pattern == 'sustained')
    x = xycoords.loc[df[idx].elec]['x_2d']
    y = xycoords.loc[df[idx].elec]['y_2d']
    tmp = ax.scatter(x, y, facecolors = 'goldenrod', edgecolor = None,  s = 200)
    sustained = df[idx][param].mean()

    D = df[df.pattern == 'D'][param].mean()
    
    textstr = 'D = %.1f\nS/SR = %.1f\nR = %.1f\nsustained = %.1f'%(D, S, R, sustained)
    props = dict(boxstyle='round', facecolor ='white')

    ax.text(0.7, 0.1, textstr, transform=ax.transAxes, fontsize=16, weight = 'bold', verticalalignment='top', bbox=props, color = 'black')
    
    filename = os.path.join(saveDir, '_'.join([subj, task, param]))
    if normalize:
        filename = '_'.join([filename, 'norm'])

    plt.savefig(filename + '.png')
    plt.close()
```
