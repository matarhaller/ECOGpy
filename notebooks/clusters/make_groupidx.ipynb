```python
import os
import pandas as pd
import numpy as np
import cPickle as pickle
```

```python
reconlist = '/home/knight/matar/MATLAB/DATA/Avgusta/PCA/reconlist.csv'
reconlist = pd.read_csv(reconlist)
```

```python
SJdir = '/home/knight/matar/MATLAB/DATA/Avgusta/'

filename = os.path.join(SJdir, 'PCA', 'hclust', 'groupidx_hclust.csv')
df = pd.DataFrame.from_csv(filename)
df = df.reset_index()

df.rename(columns={'active_elec':'active_elecs'}, inplace=True)
```

```python
df['active_cluster_stim'] = True
df['active_cluster_resp'] = True
```

```python
for x in reconlist.values:
    s_t, reconfile, _, _, _ = x
    subj, task = s_t.split('_')
    
    if isinstance(reconfile, float):
        continue #no reconfile, blank - skip subj
    
    #pull out significant clusters - STIMULUS LOCKED
    filename = os.path.join(SJdir,'PCA', 'ShadePlots_hclust', ''.join([s_t, '.csv']))
   
    df_st = pd.read_csv(filename)
    
    inactive = df_st.loc[((df_st.start_idx==0) & (df_st.end_idx ==0)), 'cluster']
    if len(inactive)>0:
        for i in inactive.values:
            df.loc[(df.subj== subj) & (df.task == task) & (df.group == i), 'active_cluster_stim']= False
    
    
    #pull out significant clusters - RESPONSE LOCKED
    filename = os.path.join(SJdir,'PCA', 'ShadePlots_hclust', 'resplocked_all', ''.join([s_t, '.csv']))
   
    df_st = pd.read_csv(filename)
    
    inactive = df_st.loc[((df_st.start_idx==0) & (df_st.end_idx ==0)), 'cluster']
    
    if len(inactive>0):
        for i in inactive.values:
            df.loc[(df.subj== subj) & (df.task == task) & (df.group == i), 'active_cluster_resp']= False
    
```

```python
#dropped clusters
df.loc[(df.active_cluster_stim == False) & (df.active_cluster_resp == False), ['subj', 'task']]
```

```python
filename = os.path.join(SJdir, 'PCA', 'duration_dict_500', 'groupidx_activeclusters_hclust_withduration_thresh15_maxRTlocked_withcriteria_ROI.csv')
df_full = pd.read_csv(filename)
```
