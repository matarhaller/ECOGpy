## plot single trial traces for an electrode
* mark significance window

```python
subj, task, elec = ('GP15', 'EmoGen', 40)

SJdir = '/home/knight/matar/MATLAB/DATA/Avgusta/'

filename = os.path.join(SJdir,'PCA', 'Stats', 'single_electrode_windows_withdesignation_EDITED_dropped_withROI.csv')
df = pd.read_csv(filename)

#load data
filename = os.path.join(SJdir, 'Subjs', subj, task, 'HG_elecMTX_percent_unsmoothed.mat')
data_dict = loadmat.loadmat(filename)

active_elecs, Params, srate, RT, data_all = [data_dict.get(k) for k in ['active_elecs','Params','srate','RTs','data_percent']]
bl_st = Params['bl_st']
bl_st = bl_st/1000*srate

RT = RT + round(abs(bl_st)) #RTs are calculated from stim onset, need to account for bl in HG_elecMTX_percent

row = df[(df.subj == subj) & (df.task == task) & (df.elec == elec)]
_, subj, task, cluster, pattern, elec, start_idx, end_idx, start_idx_resp, end_idx_resp, _, _ = row.values[0]

eidx = np.in1d(active_elecs, elec)
data = data_all[eidx,:,:].squeeze()

for i in range(data.shape[0]):
    f,ax = plt.subplots(figsize = (15,5)) #(15,5) good size for zoomed in
    
    ax.plot(np.arange(round(bl_st), data.shape[1]-abs(round(bl_st))), data[i,:], label = None)
    ax.axvline(start_idx, color = 'r', label = start_idx)
    ax.axvline(RT[i]+end_idx_resp, color = 'r', label = int(RT[i]+end_idx_resp))
    
    ax.set_title('%s %s %i - %s - trial %i - zoom' %(subj, task, elec, pattern, i))
    ax.legend()
    #filename = os.path.join(SJdir, 'PCA','Stats','latencies', 'static','singletrials', '%s_%s_e%i' %(subj, task, elec), 'unsmoothed', '%s_%s_e%i_%i.png' %(subj, task, elec, i))
    
    ax.set_xlim(0, 2000/1000*srate) #if zooming in
    filename = os.path.join(SJdir, 'PCA','Stats','latencies', 'static','singletrials', '%s_%s_e%i' %(subj, task, elec), 'unsmoothed', '%s_%s_e%i_%i_zoom.png' %(subj, task, elec, i))

    plt.savefig(filename)
    plt.close()

#plt.plot(data.T)
#plt.plot(data.mean(axis = 0), 'k',lw = 4)
```
