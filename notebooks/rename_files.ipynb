```python
import os
import pandas as pd
import shutil
```

```python
SJdir = '/home/knight/matar/MATLAB/DATA/Avgusta'

filename = os.path.join(SJdir, 'PCA', 'duration_dict_500_FINAL', 'groupidx_activeclusters_hclust_withduration_maxRTlocked_withcriteria_labeled.csv')

df = pd.read_csv(filename)
df=  df[['subj','task','group','active_elecs','all_criteria_passed']]
```

```python
for row in df.itertuples():
    _, subj, task, group, elec, duration = row
    filename = os.path.join(SJdir, 'PCA','ShadePlots_allelecs', 'images', ''.join([subj,'_',task, '_e', str(elec), '_bigwindow.png']))
    if duration:
        filename_new = os.path.join(SJdir, 'PCA','ShadePlots_allelecs', 'images', 'duration', ''.join([subj,'_',task, '_c', str(group), '_e', str(elec), '.png']))
    else:
        filename_new = os.path.join(SJdir, 'PCA','ShadePlots_allelecs', 'images', ''.join([subj,'_',task, '_c', str(group), '_e', str(elec), '.png']))
    
    if os.path.exists(filename):
        shutil.move(filename, filename_new)
```
