```python
fs = 250
T = 3
nsamples = T * fs
t = np.linspace(0, T, nsamples, endpoint=False)
```

```python
y = 2*sin(2*pi*t)
```

```python
y_noise = y +  np.random.random_sample(y.shape)
y_noise -= np.random.random_sample(y.shape)
```

```python
f, ax = plt.subplots()
plt.plot(t, y, zorder = 1, linewidth = 5, color = 'black')
plt.plot(t, y_noise, 'blue', zorder = 0)
plt.plot(t, np.tile(2.3, t.shape), 'r', linewidth = 3)

ax.patch.set_facecolor('white')
ax.autoscale(tight=False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()
ax.set_xlabel('Time')
ax.set_ylabel('Signal Strength')

ax.tick_params(labelbottom='off', labelleft = 'off')

ax.set_ylim([-3,4])
filename = os.path.join(SJdir, 'PCA','figs_for_Bob', 'stochastic_resonance.png')
#plt.savefig(filename, dpi = 100)
```
